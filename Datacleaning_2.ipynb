```python
import pandas as pd

df = pd.DataFrame({'Pclass': [1, 2, 3], 'Sex': ['male', 'female', 'male'], 'Embarked': ['C', 'S', 'Q']})
print(df)

dummies = []
cols = ['Pclass', 'Sex', 'Embarked']
for col in cols:
  dummies.append(pd.get_dummies(df[col]))
```
